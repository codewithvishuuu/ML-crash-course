```python
import numpy as np

# Broadcasting (NumPy magic)
scores = np.array([12, 32, 65, 45, 95, 15, 75, 37])

bonus = scores + 5
print(bonus)

#Instead of looping, we can directly add a scalar (single value) to a numpy array.
# Numpy automatically broadcasts the value to match the array's shape.
arr_2D = np.array([[80, 90, 85],
                   [70, 60, 50],
                   [88, 76, 95]])

grace = np.array([2, 3, 4]) # add every column value !
result = arr_2D + grace
print(result)
```

```python
# Vectorization & Handling Missing Values ?

arr = np.arange(1, 1000001)
squared = arr **2
print(squared)

# Vectorization means replacing lopps with array opertions.
# This makes code shorter, cleaner and thousands of time faster.

# Handling Missing Values ?

# In real world datasets (Excel, Csv, Database), we often have missing values (nan)
#Numpy gives us tools to direct and handle them
data = np.array([10, 30, np.nan, 70, np.nan, 10])
print(data)

# Detect missing value!
print(np.isnan(data))

#Count missing values!
print(np.isnan(data).sum())

#Replace nan with mean
mean = np.nanmean(data)
clean = np.where(np.isnan(data), mean, data)
clean
```
