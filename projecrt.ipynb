```python
import numpy as np
from matplotlib.image import imread
import matplotlib.pyplot as plt
```

```python
image = imread("ai-generated.jpg")
print("image shape ", image.shape)
```

```python
# Extract RGB channels :--

R = image[ :, :, 0]
G = image[ :, :, 1]
B = image[ :, :, 2]


# Convert to Grayscale 
#Numpy concept here : Broadcasting & elements- waise operations.
grayscale = 0.2989 * R + 0.5870 * G + 0.1140 * B
print("grayscale Shape :", grayscale.shape)


# Apply Threshold
# syntax : np.where(condition, value_ifTrue, value_ifFalse)
threshold_value = 128
threshold_value = np.where(grayscale > threshold_value, 255, 0)

```

```python
# Display image
plt.figure(figsize=(12, 4))

ax1 = plt.subplot(1, 3, 1)
ax1.set_title("Original")
ax1.imshow(image.astype(np.uint8))

plt.subplot(1, 3, 2)
plt.title("Grayscale")
plt.imshow(grayscale, cmap="gray")

plt.subplot(1, 3, 3)
plt.title("Threshold")
plt.imshow(threshold_value, cmap="gray")

plt.show()
```

```python
# Display images using Matplotlib
# -------------------------------

# plt.subplot(1, 3, n)
# Iska use ek hi row me 3 images display karne ke liye hota hai

# astype(np.uint8)
# Pixel values ko integer (0255) range me convert karta hai

# Create a new figure / canvas for plotting images
plt.figure(figsize=(12, 4))

# figsize = (width, height)
# Figure ka size inches me define karta hai


# Subplot 1 : Original Image
# -------------------------------

# Divide figure into 1 row, 3 columns
# and select the 1st position
plt.subplot(1, 3, 1)

# Set title for first image
plt.title("Original")

# Display original image
plt.imshow(image.astype(np.uint8))


# Subplot 2 : Grayscale Image
# -------------------------------

# Select the 2nd position
plt.subplot(1, 3, 2)

# Set title
plt.title("Grayscale")

# cmap="gray"
# Matplotlib ko grayscale me image dikhane ke liye bolta hai
plt.imshow(grayscale, cmap="gray")


# Subplot 3 : Threshold Image
# -------------------------------

# Select the 3rd position
plt.subplot(1, 3, 3)

# Set title
plt.title("Threshold")

# Display threshold image in grayscale
plt.imshow(threshold_value, cmap="gray")


# Adjust layout
# -------------------------------

# Automatically spacing adjust karta hai
# taaki titles aur images overlap na kare
plt.tight_layout()

# Show all plots
plt.show()

```
