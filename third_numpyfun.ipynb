```python
import numpy as np

arr = np.array([3, 5, 9, 7, 6, 8])
matrix = np.array([[3, 2, 1],
                   [6, 8, 9],
                   [5, 4, 7]])
```

```python
s = np.sum(arr)
print("Sum of array :", s)

s = np.sum(matrix)
print("Sum of matrix :", s)
s = np.sum(matrix, axis=0) #column
print(s)
s = np.sum(matrix, axis=1) #ROW
print(s)
```

```python
p = np.prod(arr)
print("Product of array :", p)
print(matrix)
```

```python
mi = np.min(arr)
print("minimum value of array :", mi)
ma = np.max(arr)
print("maximum value of array :", ma)
```

```python
print(ma-mi) #Range of array
p = np.ptp(arr)
print(p)
```

```python
cs = np.cumsum(arr)
print(cs)

cp = np.cumprod(arr)
print(cp)
```

```python
a = np.all(arr>3)
print(a)

an = np.any(arr>3)
print(an)
```

```python
cnz =np.count_nonzero(arr)
print(cnz)
```

```python
am = np.argmax(arr)
print(am)

am = np.argmin(arr)
print(am)
```

```python
s = np.size(arr)
print(s)

m = np.size(matrix)
print(s)
```

```python
nd = np.ndim(arr)
print(nd)

nd = np.ndim(matrix)
print(nd)
```

```python
s = np.shape(matrix)
print(s)

s = np.shape(arr)
print(s)
```
